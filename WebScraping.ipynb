```python

import pandas as pd
import requests
from bs4 import BeautifulSoup
import json

# get infection data
url_infected = 'https://coronadashboard.rijksoverheid.nl/landelijk/positief-geteste-mensen'
site_infected  = requests.get(url_infected).text
soup = BeautifulSoup(site_infected, 'html5lib')
soup.prettify()
data_infected = soup.find_all('script')[19].string
b = json.loads(data_infected)
json_infected = b['props']['pageProps']['selectedNlData']['tested_ggd']['values']
df_infected = pd.DataFrame(json_infected)
df_infected['Date']= pd.to_datetime(df_infected['date_unix'], origin = 'unix', unit = 's').dt.date
df_infected


```

```python
url_ic = 'https://coronadashboard.rijksoverheid.nl/landelijk/intensive-care-opnames'
site_ic  = requests.get(url_ic).text
soup = BeautifulSoup(site_ic, 'html5lib')
soup.prettify()
data_ic = soup.find_all('script')[19].string
c = json.loads(data_ic)
json_ic = c['props']['pageProps']['selectedNlData']['intensive_care_lcps']['values']
df_ic = pd.DataFrame(json_ic)
df_ic['Date']= pd.to_datetime(df_ic['date_unix'], origin = 'unix', unit = 's').dt.date
df_ic

```

```python
url_hospital = 'https://coronadashboard.rijksoverheid.nl/landelijk/ziekenhuis-opnames'
site_hospital  = requests.get(url_hospital).text
soup = BeautifulSoup(site_hospital, 'html5lib')
soup.prettify()
data_hospital = soup.find_all('script')[20].string
d = json.loads(data_hospital)
json_hospital = d['props']['pageProps']['selectedNlData']['hospital_lcps']['values']
df_hospital = pd.DataFrame(json_hospital)
df_hospital['Date']= pd.to_datetime(df_hospital['date_unix'], origin = 'unix', unit = 's').dt.date
df_hospital
```

```python
import matplotlib.pyplot as plt
```

```python
ax= df_infected.plot(x="Date", y=["infected_moving_average"])
ax = df_hospital.plot(x="Date", y=["beds_occupied_covid"], ax=ax)
ax = df_ic.plot(x="Date", y=["beds_occupied_covid"], ax=ax)
labels = ["nr of positive test", "ic beds", "hospital beds"]
ax.legend(labels = labels)

```
